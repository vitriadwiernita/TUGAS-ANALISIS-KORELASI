```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr

# Kalau pakai Colab, aktifkan upload dulu
from google.colab import files
uploaded = files.upload()

# Baca file CSV
data = pd.read_csv("data_tidur_konsentrasi.csv", sep=";")

# Bersihkan nama kolom (biar nggak kena KeyError lagi)
data.columns = data.columns.str.strip()

# Ambil masing-masing variabel
durasi_tidur = data["Durasi_Tidur_Jam"]
skor_konsentrasi = data["Skor_Konsentrasi"]

print("Data Durasi Tidur:")
print(durasi_tidur.values)

print("\nData Skor Konsentrasi:")
print(skor_konsentrasi.values)

# Hitung korelasi Pearson
korelasi_pearson, p_value = pearsonr(durasi_tidur, skor_konsentrasi)

print(f"\nKoefisien Korelasi Pearson: {korelasi_pearson:.3f}")
print(f"P-value: {p_value:.5f}")

# Interpretasi hasil
if korelasi_pearson > 0.8:
    interpretasi = "Korelasi sangat kuat positif"
elif korelasi_pearson > 0.6:
    interpretasi = "Korelasi kuat positif"
elif korelasi_pearson > 0.4:
    interpretasi = "Korelasi sedang positif"
elif korelasi_pearson > 0.2:
    interpretasi = "Korelasi lemah positif"
elif korelasi_pearson > 0:
    interpretasi = "Korelasi sangat lemah positif"
elif korelasi_pearson < -0.8:
    interpretasi = "Korelasi sangat kuat negatif"
elif korelasi_pearson < -0.6:
    interpretasi = "Korelasi kuat negatif"
elif korelasi_pearson < -0.4:
    interpretasi = "Korelasi sedang negatif"
elif korelasi_pearson < -0.2:
    interpretasi = "Korelasi lemah negatif"
elif korelasi_pearson < 0:
    interpretasi = "Korelasi sangat lemah negatif"
else:
    interpretasi = "Tidak ada korelasi linier (atau sangat mendekati nol)"

print(f"Interpretasi: {interpretasi}")
```

```python
import matplotlib.pyplot as plt

# Buat scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(durasi_tidur, skor_konsentrasi)

# Tambahkan label dan judul
plt.xlabel('Durasi Tidur (Jam)')
plt.ylabel('Skor Konsentrasi')
plt.title('Scatterplot Korelasi antara Durasi Tidur dan Skor Konsentrasi')
plt.grid(True)

# Tampilkan plot
plt.show()
```

```python
import pandas as pd
from scipy.stats import pearsonr

# Kalau pakai Colab
from google.colab import files
uploaded = files.upload()

# Baca CSV (ingat Excel Indonesia pakai ;)
data = pd.read_csv("data_tidur_konsentrasi.csv", sep=";")

# Ambil variabel
durasi_tidur = data["Durasi_Tidur_Jam"]
skor_konsentrasi = data["Skor_Konsentrasi"]

# Hitung korelasi dan p-value
korelasi_pearson, p_value = pearsonr(durasi_tidur, skor_konsentrasi)

# Tingkat signifikansi
alpha = 0.05

print(f"Koefisien Korelasi Pearson: {korelasi_pearson:.3f}")
print(f"P-value: {p_value:.5f}")
print(f"Tingkat Signifikansi (alpha): {alpha}")

# Uji hipotesis
if p_value < alpha:
    keputusan = "Tolak Hipotesis Nol (H0)"
    interpretasi_uji = "Ada hubungan linier yang signifikan antara Durasi Tidur dan Skor Konsentrasi."
else:
    keputusan = "Gagal Menolak Hipotesis Nol (H0)"
    interpretasi_uji = "Tidak terdapat hubungan linier yang signifikan antara Durasi Tidur dan Skor Konsentrasi."

print(f"\nKeputusan: {keputusan}")
print(f"Interpretasi Uji: {interpretasi_uji}")
```
